```python
data_artist = 'Big Thief'
data_album = 'Two Hands'
genius = lyricsgenius.Genius(geniusclientaccesstoken)
artist = genius.search_artist(data_artist, max_songs=3)
for spot_song in data[data_artist]['Albums'][data_album]['tracklist'].keys():
    #print(f'Spotify Song Name: {spot_song}')
    song = artist.song(f'{spot_song}')
    #print(f'type of song: {type(song)}')
    if song != None:
        lyrics = song.lyrics
        data[data_artist]['Albums'][data_album]['tracklist'][spot_song]['lyrics'] = lyrics
    else:
        data[data_artist]['Albums'][data_album]['tracklist'][spot_song]['lyrics'] = 'is_instrumental_or_did_not_find'
        print(f'Did not find lyrics for {spot_song}')
```
